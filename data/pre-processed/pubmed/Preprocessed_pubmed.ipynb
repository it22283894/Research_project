```python
import pandas as pd
import numpy as np
import ast  # For safely parsing string lists

df=pd.read_csv('pubmed_raw_data.csv')
```

```python
print("Original shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df
```

```python
print("\nNull values per column:")
print(df.isnull().sum())
```

```python
critical_cols = ['authors', 'title','diseases_mentioned']  # Add more if needed, e.g., 'title'
df = df.dropna(subset=critical_cols)
```

```python
print(df.isnull().sum())
```

```python
text_cols = [ 'mesh_disease_terms']
df[text_cols] = df[text_cols].fillna('')
```

```python
print(df.isnull().sum())
```

```python
str_cols = df.select_dtypes(include='object').columns
str_cols
```

```python
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df[str_cols] = df[str_cols].apply(lambda x: x.str.replace(r'\s+', ' ', regex=True) if x.dtype == "object" else x)
```

```python
print("\nFinal shape:", df.shape)
print("\nFinal null check:")
print(df.isnull().sum())
print("\nFirst 5 rows after cleaning:")
df.head()
```

```python
output_file = 'Preprocessed_pubmed_data.csv'
df.to_csv(output_file, index=False)
print(f"\nCleaned data saved to {output_file}")
```
