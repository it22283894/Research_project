```python
from neo4j import GraphDatabase

URI = "bolt://127.0.0.1:7687"
USERNAME = "neo4j"
PASSWORD = "sakuni200211"
```

```python
driver = GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD))
```

```python
def run_query(query):
    with driver.session() as session:
        result = session.run(query)
        return [dict(record) for record in result]
```

```python
print("1. All Nodes:")
nodes = run_query("MATCH (n) RETURN labels(n), count(n) AS count ORDER BY count DESC")
for n in nodes:
    print(f"  {n['labels(n)']}: {n['count']}")
```

```python
print("\n2. Relationships:")
rels = run_query("MATCH ()-[r]->() RETURN type(r) AS type, count(r) AS count ORDER BY count DESC")
for r in rels:
    print(f"  {r['type']}: {r['count']}")
```

```python
print("\n3. Avg Rels per Food:")
avg = run_query("MATCH (f:Food) OPTIONAL MATCH (f)-[r]-(o) WITH f, count(r) AS degree RETURN avg(degree) AS avg_degree")[0]['avg_degree']
print(f"  Avg: {avg:.2f} (expected ~3-5)")
```

```python
print("\n4. Sample Paths:")
samples = run_query("MATCH (f:Food)-[:CONTAINS_INGREDIENT]->(i:Ingredient) RETURN f.name, i.name LIMIT 5")
for s in samples:
    print(f"  {s['f.name']}: {s['i.name']}")
```

```python
print("\n5. Unique Entities:")
unique_ings = run_query("MATCH (i:Ingredient) RETURN count(DISTINCT i.name) AS unique_ings")[0]['unique_ings']
unique_brands = run_query("MATCH (b:Brand) RETURN count(DISTINCT b.name) AS unique_brands")[0]['unique_brands']
print(f"  Unique Ingredients: {unique_ings}")
print(f"  Unique Brands: {unique_brands}")
```

```python
driver.close()
print("\nValidation done!")
```
