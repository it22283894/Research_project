```python
import pandas as pd
```

```python
import os
```

```python
df = pd.read_csv("D:/Downloads/Research Project/Data/processed/cleaned_usda_data.csv")
```

```python
print("Loaded", len(df), "foods")
```

```python
triples = []
count = 0
for i, row in df.iterrows():
    count += 1
    food_id = "food_" + str(row["fdc_id"])
    
    triples.append([food_id, "has_name", str(row["description"])])
    
    if pd.notna(row["brand_owner"]):
        triples.append([food_id, "has_brand", str(row["brand_owner"])])
    
    if pd.notna(row["ingredient_list"]):
        ingredients = eval(row["ingredient_list"])  # convert string to list
        for ing in ingredients:
            triples.append([food_id, "contains_ingredient", ing])
    
    if count % 30000 == 0:
        print("Processed", count, "foods...")
```

```python
os.makedirs("D:/Downloads/Research Project/Data/processed/kg", exist_ok=True)
```

```python
all_triples = pd.DataFrame(triples, columns=["Subject", "Predicate", "Object"])
```

```python
part = 1
for i in range(0, len(all_triples), 5000000):
    chunk = all_triples[i:i+5000000]
    chunk.to_csv(f"D:/Downloads/Research Project/Data/processed/kg/triples_part_{part}.csv", index=False)
    print("Saved part", part)
    part += 1

print("FINISHED! Total triples =", len(all_triples))
```
