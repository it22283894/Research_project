```python
branded["ingredients"] = branded["ingredients"].str.upper()
branded["ingredients"] = branded["ingredients"].str.replace("INGREDIENTS:", "")
branded["ingredients"] = branded["ingredients"].str.replace("CONTAINS.*", "", regex=True)
branded["ingredients"] = branded["ingredients"].str.replace(r"\(.*?\)", "", regex=True)
```

```python
branded["ingredient_list"] = branded["ingredients"].str.split(",")
```
